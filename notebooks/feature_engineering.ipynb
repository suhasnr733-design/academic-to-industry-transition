```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Load data from Sridharshan's pipeline
df = pd.read_csv('../data/processed/student_data_cleaned.csv')
print(f"Dataset shape: {df.shape}")

# Feature engineering
def create_features(df):
    df_copy = df.copy()
    
    # Skill-based features
    df_copy['skill_count'] = df_copy['skills'].apply(lambda x: len(x) if isinstance(x, list) else 0)
    df_copy['skill_diversity'] = df_copy['skills'].apply(lambda x: len(set(x)) if isinstance(x, list) else 0)
    
    # Experience features
    df_copy['total_experience'] = df_copy['internship_months'] + df_copy['projects'] * 2
    
    # Academic features
    df_copy['cgpa_normalized'] = df_copy['cgpa'] / 10
    
    # Certification features
    df_copy['certification_score'] = df_copy['certifications'] * 2 + df_copy['workshops'] * 1
    
    # Interaction features
    df_copy['skill_cgpa_ratio'] = df_copy['skill_count'] / (df_copy['cgpa'] + 1)
    df_copy['exp_skill_ratio'] = df_copy['total_experience'] / (df_copy['skill_count'] + 1)
    
    # Department encoding
    le = LabelEncoder()
    df_copy['department_encoded'] = le.fit_transform(df_copy['department'])
    
    return df_copy

df_engineered = create_features(df)
df_engineered.to_csv('../data/processed/student_data_engineered.csv', index=False)
print("✅ Feature engineering complete!")
```
